# use beautiful soup to parse and extract information 

```python
!pip install beautifulsoup4 --upgrade 
```

```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page_content, "html.parser")
```

```python
doc
```

```python
selection_class = "f3 lh-condensed mb-0 mt-1 Link--primary"
topic_title_tags = doc.find_all('p',{"class" : selection_class })
```

```python
len(topic_title_tags)
```

```python
topic_title_tags
```

```python
topic_title_tags[:5]
```

```python
desc_selector = "f5 color-fg-muted mb-0 mt-1"

topic_description_tags = doc.find_all('p',{"class" : desc_selector })
topic_description_tags[:5]
```

```python
topic_link_tags= doc.find_all('a',{'class':'d-flex no-underline'})
```

```python
len(topic_link_tags)
```

```python
topic_link_tags[0]['href']
```

```python
topic0_url="https://github.com" + topic_link_tags[0]['href']
```

```python
print(topic0_url)
```

```python
topic_titles=[]
topic_description=[]
topic_urls=[]
base_url = "https://github.com"

for tag in topic_title_tags : 
    topic_titles.append(tag.text)
    
for tag2 in topic_description_tags : 
    topic_description.append(tag2.text.strip())
    
for tag3 in topic_link_tags : 
    topic_urls.append(base_url + tag3['href'])
    
print(topic_titles)
print("\n")
print(topic_description)
print("\n")
print(topic_urls)

```

```python
!pip install  pandas 
```

```python
topics_dict= { 
'title':topic_titles , 
'description': topic_description , 
'url' : topic_urls
}
```

```python
import pandas as pd 
topic_df=pd.DataFrame(topics_dict)
```

```python
topic_df
```

```python
star_tags = topic_doc.find_all('a',{'class':'social-count js-social-count'})
len(star_tags)
```

```python
star_tags[0].text.strip()
```

```python
def parse_star_count (stars_str) : 
    stars_str = stars_str.strip()
    if stars_str[-1]=='k' : 
        return int( float(stars_str[:-1]) * 1000)
    return int(stars_str)
    
```

```python
parse_star_count(star_tags[0].text.strip())
```

```python
def get_repo_info(h3_tag  , star_tag) :
    #return all the required information about the repository 
    a_tags = h3_tag.find_all('a') 
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href']
    stars = parse_star_count(star_tag.text.strip())
    return username , repo_name , stars ,repo_url

    
    
```

```python
get_repo_info(repo_tags[0],star_tags[0])
```

```python
topic_repos_dict = {
    'username' : [],
    'repo_name' : [] , 
    'stars':[] , 
    'repo_url' :[]
}    
for i in range(len(repo_tags)):
    repo_info= get_repo_info(repo_tags[i],star_tags[i])
    topic_repos_dict['username'].append(repo_info[0])
    topic_repos_dict['repo_name'].append(repo_info[1])
    topic_repos_dict['stars'].append(repo_info[2])
    topic_repos_dict['repo_url'].append(repo_info[3])
```

```python
topic_repos_df = pd.DataFrame(topic_repos_dict)
```
